```python
import pandas as pd
df = pd.read_csv('parsedStave.csv')
df
```

```python
# Extracting milliseconds from Milliseconds today which we will end up using in a datetime attribute 


df['millisecondsToday'].astype(str)


df['millisecondsToday'] = df['millisecondsToday'].str.strip('[]').astype(int)

df['milliseconds'] = df['millisecondsToday'] % 1000

# df['milliseconds'] = df['millisecondsToday'].astype(str)
df['milliseconds'] = df['milliseconds'].astype(str)
```

```python
# Goal is to clean the following attributes posFixDay,posFixHours,posFixMinutes,postFixSeconds
# ['posFixDay','posFixHours','posFixMinutes','postFixSeconds']

df['posFixMonth'] = df['posFixDay'].str.strip('[]').astype(int) 
df['posFixMonth'] = df['posFixMonth'] // 25
# i did the floor function of 25 because the data is in february and i didnt feel like getting a ceiling function of 31

df['posFixMonth'] = df['posFixMonth'].astype(str)
df['posFixMonth'] = df.posFixMonth.str.strip()
df['posFixMonth'] = df.posFixMonth.str.pad(width = 2, side = 'left', fillchar = '0')


df['posFixDay'] = df['posFixDay'].str.strip('[]').astype(int) % 31
df['posFixDay'] = df['posFixDay'].astype(str)

df['posFixHours'] = df['posFixHours'].str.strip('[]').astype(str)

df['posFixMinutes'] = df['posFixMinutes'].str.strip('[]').astype(str)

df['postFixSeconds'] = df['postFixSeconds'].str.strip('[]').astype(str)

df['posFixYear'] = df['posFixYear'].str.strip('[]').astype(str)

df['dateTime'] = df['posFixYear'] +"/" + df['posFixMonth'] +"/"+ df['posFixDay'] +" " + df['posFixHours'] + ":" + df['posFixMinutes'] + ":" + df['postFixSeconds'] + ":" + df['milliseconds']

# 2021/02/22 15:33:29:188
df['dateTime'] =  pd.to_datetime(df['dateTime'], format='%Y/%m/%d %H:%M:%S:%f') 


df = df.set_index(['dateTime'])
df =  df[['altitude' , 'pingNum',]]

df['altitude'] = df['altitude'].str.strip('[]').astype(int)

df['pingNum'] = df['pingNum'].str.strip('[]').astype(int)
df
# print(df['dateTime'] )
```

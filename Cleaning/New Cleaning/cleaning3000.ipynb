```python
import pandas as pd
df = pd.read_csv('3001.csv')
df
```

```python



# df['PingNum'].astype(str)
# df['Channel'].astype(str)
# df['Seconds'].astype(str)
# df['Nanoseconds'].astype(str)
# df['DelayIndex'].astype(str)
df['seconds'] = df['seconds'].astype(int)
df['nano'] = df['nano'].astype(str)



# df['PingKey']= df.PingKey.str.strip('[]').astype(int)
# df['Channel']= df.Channel.str.strip('[]').astype(int)
# df['Seconds']= df.Seconds.str.strip('[]').astype(int)
# df['Angle']= df.Angle.str.strip('[]').astype('float32')
# df['DelayIndex']= df.DelayIndex.str.strip('[]').astype(int)

df['nano']= df.nano.str[:3]
df['nano'] = '.' + df['nano']
df

# pd.to_datetime(1.547559e+09, unit='s', origin='unix') 

# 2021-02-22 
df['dateTime']=pd.to_datetime(df['seconds'],  unit='s', origin='unix')
df['dateTime'] = df['dateTime'].dt.strftime('%Y-%d-%m %H:%M:%S')
df['dateTime'].astype(str)
df['dateTime'] = df['dateTime'] + df['nano']
# df['dateTime'] =  pd.to_datetime(df['dateTime'], format='%Y-%m/%d %H:%M:%S:%f') 
df['dateTime'] =  pd.to_datetime(df['dateTime'], format='%Y-%d-%m %H:%M:%S.%f') 

df = df[['channelNum', 'seqNum','flags','headings','heaves','pitch','roll','yaw','dateTime']]
df.to_csv("3000.csv",index=False)

# print((df['Angle'].iloc[1]))

df
```
